```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

color_dict = dict({'Conserved': 'k',
                   'Lost in egg': 'r',
                  'Gained in MDCK':'g',
                  'Gained in egg': 'b',
                  'Lost in MDCK': 'm'})
```

```python
def getPlot(egg_status):
    egg_status['Index'] = egg_status.index
    twoThree = egg_status[egg_status.GlycType == '2-3 SiaLac']
    twoSix = egg_status[egg_status.GlycType == '2-6 SiaLac']
    
    fig, ax = plt.subplots(1,2, figsize=(16,6))
    sns.scatterplot(y='Valency', x='PolSpace', data=twoThree, hue='Status', ax=ax[0], palette=color_dict, s=60)
    sns.scatterplot(y='Valency', x='PolSpace', data=twoSix, hue='Status', ax=ax[1], palette=color_dict, s=60)
    ax[0].get_legend().remove()
    ax[1].legend(fontsize=18)
    ax[0].set_ylabel("Valency", fontsize=18)
    ax[0].set_xlabel("Polymer density", fontsize=18)
    ax[0].set_xlim(-0.2,5.5)
    ax[0].xaxis.set_tick_params(labelsize=18)
    ax[0].set_ylim(40,150)
    ax[0].yaxis.set_tick_params(labelsize=18)
    ax[1].set_ylabel("Valency", fontsize=18)
    ax[1].set_xlim(-0.2,5.5)
    ax[1].set_xlabel("Polymer density", fontsize=18)
    ax[1].set_ylim(40,150)
    ax[1].xaxis.set_tick_params(labelsize=18)
    ax[1].yaxis.set_tick_params(labelsize=18)
```

```python
allDat = pd.read_csv("./All_Feb1.csv")
allDat_good = allDat[allDat.Status != 'Others']
```

```python
allDat_good2 = allDat_good[(allDat_good.Status != 'Lost in egg') & (allDat_good.Status != 'Gained in egg')]
allDat_good3 = allDat_good[allDat_good.Status != 'Gained in MDCK']
```

```python
getPlot(allDat_good2)
#plt.savefig("VennDiagram_GainedInMDCK_flipped.pdf")
```

```python
getPlot(allDat_good3)
#plt.savefig("VennDiagram_LostInEgg_flipped.pdf")
```
