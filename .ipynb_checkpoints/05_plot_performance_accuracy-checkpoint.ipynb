```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
egg = np.loadtxt("egg_all.txt")
mdck = np.loadtxt("mdck_all.txt")
```

```python
fig, ax = plt.subplots(1,1, figsize=(8,6))

ax.scatter(egg[:,0], egg[:,1], color='k', label='Egg virus')
ax.plot(egg[:,0], egg[:,1], color='k')

left, bottom, width, height = [0.52, 0.2, 0.35, 0.5]
ax2 = fig.add_axes([left, bottom, width, height])
ax2.scatter(mdck[:,0], mdck[:,1], color='r', label='MDCK virus')
ax2.plot(mdck[:,0], mdck[:,1], color='r')

ax.legend(fontsize=18)
ax.set_xticks(np.arange(0, 2001, 500))
ax.xaxis.set_tick_params(labelsize=18)
ax.set_yticks(np.arange(0.6, 0.8, 0.1))
ax.yaxis.set_tick_params(labelsize=18)
ax.set_xlabel("Number of iterations", fontsize=18)
ax.set_ylabel("Prediction accuracy", fontsize=18)

ax2.legend(fontsize=18)
ax2.set_xticks(np.arange(0, 201, 50))
ax2.xaxis.set_tick_params(labelsize=15)
ax2.set_yticks(np.arange(0.25, 0.8, 0.25))
ax2.yaxis.set_tick_params(labelsize=15)

plt.savefig("Plots/PerformanceAccuracy.pdf")
```
