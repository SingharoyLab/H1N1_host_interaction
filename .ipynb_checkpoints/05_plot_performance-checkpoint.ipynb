```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
egg = np.loadtxt("egg_all.txt")
mdck = np.loadtxt("mdck_all.txt")
```

```python
fig, ax = plt.subplots(1,2, figsize=(16,6))

ax[0].scatter(egg[:,0], egg[:,1], color='k', label='Accuracy')
ax[0].plot(egg[:,0], egg[:,1], color='k')
ax[0].scatter(egg[:,0], egg[:,2], color='r', label='Precision')
ax[0].plot(egg[:,0], egg[:,2], color='r')
ax[0].scatter(egg[:,0], egg[:,3], color='b', label='Recall')
ax[0].plot(egg[:,0], egg[:,3], color='b')
ax[0].scatter(egg[:,0], egg[:,4], color='g', label='F-1 score')
ax[0].plot(egg[:,0], egg[:,4], color='g')

#ax[0].legend(fontsize=18)
ax[0].set_xticks(np.arange(0, 2001, 500))
ax[0].xaxis.set_tick_params(labelsize=18)
ax[0].set_yticks(np.arange(0.6, 1.1, 0.2))
ax[0].yaxis.set_tick_params(labelsize=18)
ax[0].set_xlabel("Number of iterations", fontsize=18)
ax[0].set_ylabel("Prediction accuracy", fontsize=18)

ax[1].scatter(mdck[:,0], mdck[:,1], color='k', label='Accuracy')
ax[1].plot(mdck[:,0], mdck[:,1], color='k')
ax[1].scatter(mdck[:,0], mdck[:,2], color='r', label='Precision')
ax[1].plot(mdck[:,0], mdck[:,2], color='r')
ax[1].scatter(mdck[:,0], mdck[:,3], color='b', label='Recall')
ax[1].plot(mdck[:,0], mdck[:,3], color='b')
ax[1].scatter(mdck[:,0], mdck[:,4], color='g', label='F-1 score')
ax[1].plot(mdck[:,0], mdck[:,4], color='g')

ax[1].legend(fontsize=18)
ax[1].set_xticks(np.arange(0, 201, 50))
ax[1].xaxis.set_tick_params(labelsize=15)
ax[1].set_yticks(np.arange(0.3, 0.91, 0.3))
ax[1].yaxis.set_tick_params(labelsize=15)
ax[1].set_xlabel("Number of iterations", fontsize=18)
ax[1].set_ylabel("Prediction accuracy", fontsize=18)

plt.savefig("Plots/Performance_vs_all.pdf")
```
