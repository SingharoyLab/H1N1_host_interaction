```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
correct = np.loadtxt("Results/egg_convergence.txt")
correct_final = correct[-1][1:]
```

```python
random = np.loadtxt("../Results_Jun2021_randomSplit/egg_all.txt")
random_final = random[-1][1:]
```

```python
plt.figure(figsize=(8,6))

plt.bar(np.arange(0,8,2), correct_final, width=0.5, label='Split individually')
plt.bar(np.arange(0,8,2)+0.5, random_final, width=0.5, label='Split across experiments')

plt.xticks(np.arange(0,8,2))
plt.gca().set_xticklabels(['Accuracy', 'Precision', 'Recall', 'F1 score'], rotation=45, fontsize=16)
plt.yticks(np.arange(0.0,1.1,0.25), fontsize=16)
plt.ylim(0,1.15)
plt.ylabel("Performance scores", fontsize=18)
plt.legend(fontsize=15, loc=2)

plt.gcf().subplots_adjust(bottom=0.16)
plt.savefig("Plots/Comparison.pdf")
```
